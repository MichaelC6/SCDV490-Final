```python
import folium
from util.preProcessing import readXMLOptions
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
datadir = os.path.join(os.getcwd(), 'sampleData')
resultfile = os.path.join(datadir, 'test.xml')

result = readXMLOptions(resultfile,'way')
```

```python
# make plot of every single node we have
#print(result.nodes)
coords = {'id':[], 'lat':[], 'long':[]}
for n in result.nodes:
    coords['id'].append(n.id)
    coords['lat'].append(n.lat)
    coords['long'].append(n.lon)
    
coords = pd.DataFrame(coords)
```

```python
plt.figure(figsize=(40,40))
plt.scatter(coords['lat'],coords['long'],marker=1)
```

```python
m = folium.Map(location=[np.mean(coords['lat']),np.mean(coords['long'])])
```

```python
for i in range(0,len(coords)):
    if i % 10000:
        folium.CircleMarker(
            radius=25,
            location=[coords['lat'][i],coords['long'][i]],
            popup="The Waterfront",
            color="crimson",
            fill=False,
        ).add_to(m)
```
